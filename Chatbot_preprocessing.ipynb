```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("amazon_co-ecommerce_sample.csv")
```

```python
data
```

```python
data.describe()
```

```python
null_columns = data.columns[data.isnull().any()]
data[null_columns].isnull().sum()
```

```python
data_1 = data.copy()
```

```python
df= pd.DataFrame(data_1)
```

```python
## Delete Index Column
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
## Droping all the columns which have high null values

df.drop(["number_of_answered_questions", "description", "items_customers_buy_after_viewing_this_item", "customer_questions_and_answers" ], axis=1, inplace=True)
```

```python
df["price"].isnull().sum()
```

```python
box = float()
df["newPrice"] = float()
newData = []
for i in df['price']:
    box = str(i).replace('£', '')
    newData.append(box)

df["newPrice"] = newData


df["newPrice"]
```

```python
df = df.reindex(columns=['uniq_id', 'product_name', 'manufacturer', 'newPrice','price',
       'number_available_in_stock', 'number_of_reviews',
       'average_review_rating', 'amazon_category_and_sub_category',
       'customers_who_bought_this_item_also_bought', 'product_information',
       'product_description', 'customer_reviews', 'sellers', 'newPrice'])
```

```python
df.drop(["price"],axis=1, inplace=True)
df["newPrice"]= df["newPrice"].fillna(0)
```

```python

df['number_available_in_stock'] = df['number_available_in_stock'].str.replace(r'\D', '')
```

```python
df["number_available_in_stock"]= df["number_available_in_stock"].fillna(0)
df["number_available_in_stock"].isnull().sum()
```

```python
df["number_available_in_stock"].astype(int)
```

```python
df["average_review_rating"]= df["average_review_rating"].fillna(0)
df["average_review_rating"].isnull().sum()
```

```python
df["number_of_reviews"]= data_1["number_of_reviews"].fillna(0)
df["number_of_reviews"].isnull().sum()
```

```python
df.drop(["sellers"], axis=1, inplace=True)
```

```python
df["customer_reviews"]= data_1["customer_reviews"].fillna(0)
df["customer_reviews"].isnull().sum()
```

```python
df["product_information"]= data_1["product_information"].fillna(0)
df["product_information"].isnull().sum()  
```

```python
df["product_description"]= data_1["product_description"].fillna(0)
df["product_description"].isnull().sum() 
```

```python
df["amazon_category_and_sub_category"]= data_1["amazon_category_and_sub_category"].fillna(0)
df["amazon_category_and_sub_category"].isnull().sum()
```

```python
df.loc[df["product_information"]==0] = df[["product_description"]]
```

```python
#df["product_information"].fillna(0)
```

```python
df.dropna(subset=['customers_who_bought_this_item_also_bought'],axis=0,inplace=True)
```

```python
df["manufacturer"]= data_1["manufacturer"].fillna(0)
df["manufacturer"].isnull().sum()
```
